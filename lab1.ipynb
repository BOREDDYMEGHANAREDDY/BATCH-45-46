```python

people=['anu','bhanu','chreey','dolly','eksha','farha']
weight=[40,32,41,76,89,56]
height=[160,143,153,167,127,123]




```

```python
import matplotlib.pyplot as plt
plt.scatter(weight,height)
plt.title("relationship between weight and height of patients")
plt.ylabel("height in meters")
plt.xlabel("weight in kilos")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("peoples weight in kgs",fontsize=16,fontstyle='italic')
plt.bar(x=people,height=weight,width=0.6,color='tomato',edgecolor='k',alpha=0.6)
plt.xlabel("people",fontsize=15)
plt.ylabel("weight(in kgs)",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='lightblue',edgecolor='blue',alpha=0.75,bins=5)
plt.title("histogram of patitent weight",fontsize=20)
plt.xlabel("weight in kilos",fontsize=15)
plt.show()
```
